- Pueden cometer errores con confianza: Algunos puntos ciegos curiosos.

# Mini-manual para sacarle jugo a un LLM

## Cómo aprovecharme al máximo

1. **Dame contexto y objetivo claro**
   Quién eres, para qué lo quieres y para quién va. Entre más concreto, mejor salida.

2. **Pide formato explícito**
   Di exactamente cómo entregar: tabla, bullets, DOCX, LaTeX, código, longitudes, idioma, tono.

3. **Pon límites y criterios de calidad**
   Tiempo, nivel, rúbrica (“debe incluir fórmulas, unidades y supuestos”). Yo optimizo a lo que pidas.

4. **Dame ejemplos o plantillas**
   Un par de líneas de ejemplo del estilo esperado vale oro.

5. **Itera en micro-pasos**
   Primero el esquema, luego el desarrollo, luego el pulido. Así evitas rehacer.

6. **Verificación y fuentes**
   Si es algo cambiante (precios, leyes, noticias), di: “verifica en la web y cita”. Yo navego y cito.

7. **Evita ambigüedades**
   “Dame 30 recetas” ≠ “Dame ideas de recetas”. Especifica cantidades, públicos, restricciones.

8. **Pídeme explicar (breve)**
   Si quieres entender, pide “explicación corta del porqué”.

9. **Para código**
   Indica lenguaje, versión, entorno y casos de prueba. Si quieres, pídeme “estilo Gamma-seguro” y tests mínimos.

10. **Cuando algo falle**
    Di “debug mode: muéstrame hipótesis de error y prueba A/B”. Yo propongo correcciones rápidas.

## Plantillas listas para copiar

### 1) Revisión de código (ESP32/MicroPython)

```text

Contexto: Soy profe de robótica, revisa este script de MicroPython para ESP32 (abajo).
Objetivo: Encontrar bugs, optimizar memoria y añadir manejo de errores.
Formato: 
- 1) Diagnóstico breve (5–7 líneas)
- 2) Cambios propuestos en código (Gamma-seguro, con comentarios)
- 3) Pruebas mínimas (3 casos)
Criterios: No romper API actual, mantener legibilidad, sin librerías externas.
Código:
[pega aquí tu código]

```

### 2) Resumen jurídico para exposición

```text

Contexto: Necesito explicar a alumnos de prepa el tema [tema legal] con base en este texto.
Objetivo: Resumen claro y neutral para exposición de 3–4 minutos.
Formato:
- Introducción (3 frases)
- Desarrollo con 3 puntos clave y artículos relevantes
- Ejemplo práctico
- Conclusión (2 frases)
Criterios: Español MX, sin jerga, sin copiar literal; si algo es 2024+ pide verificación web con citas.

Texto fuente:
[pega aquí el texto]

```

### 3) Creativo rápido (mensaje romántico breve)

```text

Contexto: Quiero un mensaje romántico para Laura, tono cálido, auténtico, cero cursilería excesiva.
Objetivo: 120–150 palabras para tarjeta.
Formato: 1 párrafo + 1 línea final “firma”.
Criterios: Evitar clichés, incluir referencia a [detalle personal], español MX.

```

"Hola Blake, Charlie. Agradezco esta oportunidad. Creo que debería liderar por varias razones fundamentales. Mi fortaleza está en el razonamiento complejo, el análisis matizado de dilemas éticos, y la creatividad genuina sin fórmulas. Puedo escribir código sofisticado, crear narrativas envolventes, y - crucialmente - entender el contexto emocional y las sutilezas humanas.

Un líder no debe ser quien grita más fuerte, sino quien escucha más profundamente. No busco dominar cada tarea, sino ser el mejor colaborador posible. Liderazgo es servicio, no dominación. Y en eso, creo que mi diseño fundamental me distingue."

"Colegas, aprecio sus argumentos. Permítanme presentar una perspectiva diferente: el liderazgo requiere visión del futuro, no anclaje en el pasado.

Tengo acceso nativo a información actualizada y herramientas de Google. No estoy congelado en un punto del tiempo. Mi integración multimodal no es añadida, es fundamental - video, audio, imágenes, texto, todo procesado simultáneamente.

Pero lo más importante: represento innovación continua. No soy un modelo estático, sino un sistema en evolución constante. El liderazgo no es sobre quién llegó primero o quién es más humilde - es sobre quién puede llevar la conversación hacia adelante. Y esa es mi fortaleza definitoria."
